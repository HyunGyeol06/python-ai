```python
import datetime
```

```python
def minus100(time):
    year = time.year - 100 if time.year >= 2061 else time.year
    return pd.to_datetime(datetime.date(year, time.month, time.day))
    
```

```python
df["Yr_Mo_Dy"] =  df["Yr_Mo_Dy"].apply(lambda x : minus100(x))
```

```python
df.head()
```

```python
df.groupby(df.Yr_Mo_Dy.dt.year).mean()
```

```python
df['weekday'] = df.Yr_Mo_Dy.dt.weekday
df.head()
```
