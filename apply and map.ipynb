```python
import pandas as pd
```

```python
DataUrl = "https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv"
```

```python
df = pd.read_csv(DataUrl)
df.head(20)
```

```python

df.shape
```

```python
df.Income_Category
```

```python
newIncome = {
    "Unknown":"N",
    "Less than $40K": "a",
    "$40K - $60K": "b",
    "$60K - $80K": "c",
    "$80K - $120K" : "d",
    "$120K +": "e"
    
}

df['newIncome'] = df['Income_Category'].map(newIncome)
df.head(5)
```

```python
newIncome = {
    "Unknown":"N",
    "Less than $40K": "a",
    "$40K - $60K": "b",
    "$60K - $80K": "c",
    "$80K - $120K" : "d",
    "$120K +": "e"
    
}

df['newIncome'] = df['Income_Category'].apply(lambda x : newIncome[x] )
df.head(5)
```

```python
import math

df['AgeState'] = df["Customer_Age"].apply(lambda x : math.trunc(x/10) )
df.head()
```

```python
df['newEduLevel'] = df['Education_Level'].apply(lambda x : 1 if x=="Graduate" else  0 )
```

```python
df.newEduLevel.value_counts()
```

```python
df['newLimit'] = df['Credit_Limit'].apply(lambda x : 1 if x>=4500 else 0)
df.head()
```

```python
def statenew(row):
    if row['Marital_Status']=="Married" and row['Card_Category'] == "Platinum":
        return 1
    else:
        return 0
```

```python
df['newState'] = df.apply(statenew, axis=1)
```

```python
df.newState.value_counts()
```

```python
def newGender(gender):
    if gender == 'M':
        return "male"
    else:
        return 'female'
```

```python
df['Gender'] = df['Gender'].apply(newGender)
df.head()
```

```python
df.Gender.value_counts()
```
