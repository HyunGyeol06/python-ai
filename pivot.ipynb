```python
import pandas as pd
import numpy as np
```

```python
Dataurl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv'
```

```python
df = pd.read_csv(Dataurl, index_col=0)
df.head()
```

```python
df.drop("Indicator",axis=1, inplace=True)
df.head()
```

```python
df['First Tooltip'] = df['First Tooltip'].apply(lambda x : x.split("[")[0])
df.head()
```

```python
df2 = df[(df['Dim1'] == "Both sexes") & (df['Period'] >= 2015)]
df2.head()
```

```python
df3 = df2.pivot(columns="Period", values=['First Tooltip'])
df3.head()
```
