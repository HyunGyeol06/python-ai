```python
import numpy as np
import pandas as pd

test = pd.read_csv("./TV_price_test.csv")
train = pd.read_csv("./TV_price_train.csv")

display(test)
display(train)
```

```python
display(test.isnull().sum())
display(train.isnull().sum())
```

```python
tv_id = train['ID']
drop_col = ["ID","판매가"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
train_input = train.drop(columns=drop_col)
test_input = test.drop(columns="ID")
train_target = train['판매가']

train_x, test_x, train_y, test_y = train_test_split(train_input, train_target, test_size = 0.3, random_state = 35)
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
gr = GradientBoostingRegressor()

lr.fit(train_x, train_y)
rf.fit(train_x, train_y)
gr.fit(train_x, train_y)
```

```python
pred_lr = lr.predict(test_x)

pred_rf = rf.predict(test_x)

pred_gr = gr.predict(test_x)
```

```python
display(pred_lr)
display(pred_rf)
display(pred_gr)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
def display_score (test_y, pred):
    mae = mean_absolute_error(test_y, pred)
    mse = mean_squared_error(test_y, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(test_y, pred)
    print("MAE: {0:.4f} , MSE: {1:.4f} , RMSE: {2:.4f} , R2: {3:.4f}".format(mae,mse,rmse,r2))
    print(" ")
```

```python
display_score(test_y, pred_lr)

display_score(test_y, pred_rf)

display_score(test_y, pred_gr)
```
