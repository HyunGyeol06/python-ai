```python
import numpy as np
import pandas as pd

test_x = pd.read_csv("./loan_prediction_X_test.csv")
train_x = pd.read_csv("./loan_prediction_X_train.csv")
train_y = pd.read_csv("./loan_prediction_Y_train.csv")

display(test_x)
display(train_x)
display(train_y)
```

```python
display(test_x.info())
display(train_x.info())
display(train_y.info())
```

```python
display(test_x.isnull().sum())
display(train_x.isnull().sum())
display(train_y.isnull().sum())
```

```python
loan_id = test_x['Loan_ID']
drop_col = ["Loan_ID","Credit_History"]
```

```python
test_x_dropped = test_x.drop(drop_col, axis=1)
train_x_dropped = train_x.drop(drop_col, axis=1)
train_y = train_x['Credit_History']

display(train_y)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
train_x_dummies = pd.get_dummies(train_x_dropped)
test_x_dummies = pd.get_dummies(test_x_dropped)[train_x_dummies.columns]


display(train_x_dummies)
display(test_x_dummies)
```

```python
train_x, test_x, train_y, test_y = train_test_split(train_x_dummies, train_y, test_size = 0.4, random_state = 66)
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
```

```python
dt.fit(train_x, train_y)

rf.fit(train_x, train_y)

gb.fit(train_x, train_y)
```

```python
import seaborn as sns
import numpy as np
%matplotlib inline

# feature importance 추출 
print("Feature importances:\n{0}".format(np.round(rf.feature_importances_, 3)))
```

```python
pred_dt = dt.predict(test_x)
pred_proba_dt = dt.predict_proba(test_x)[:,1]

pred_rf = rf.predict(test_x)
pred_proba_rf = rf.predict_proba(test_x)[:,1]

pred_gb = gb.predict(test_x)
pred_proba_gb = gb.predict_proba(test_x)[:,1]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
def display_score (test_y, pred, pred_proba):
    con = confusion_matrix(test_y, pred)
    acc = accuracy_score(test_y, pred)
    pre = precision_score(test_y, pred)
    re = recall_score(test_y, pred)
    f1 = f1_score(test_y, pred)
    roc_auc = roc_auc_score(test_y, pred_proba)
    
    print('오차 행렬')
    print(con)
    print("정확도: {0:.4f} , 정밀도: {1:.4f} , 재현율: {2:.4f} , F1: {3:.4f} , AUC: {4:.4f}".format(acc, pre, re, f1, roc_auc))
    print(" ")
```

```python
display_score(test_y, pred_dt, pred_proba_dt)
display_score(test_y, pred_rf, pred_proba_rf)
display_score(test_y, pred_gb, pred_proba_gb)
```

```python
result =  pd.DataFrame(
    {
        "ID":loan_id,
        "target": rf.predict(test_x_dummies)
    }
)
```
