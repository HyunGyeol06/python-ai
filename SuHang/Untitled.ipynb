```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./travel.csv")
```

```python
df = df.drop(["RESPOND_ID","EXAMIN_BEGIN_DE"], axis=1)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=["SEXDSTN_FLAG_CD"])
```

```python
df
```

```python
df["AGRDE_FLAG_NM"] = df['AGRDE_FLAG_NM'].str[:-1]
```

```python
df
```

```python
df_from = ["300만원 미만", "300이상500만원 미만", "500이상700만원 미만", "700만원 이상", "모름","많이 적어졌다", "약간 적어졌다", "예전과 비슷하다", "약간 커졌다", "많이 커졌다"]

df_to = [300, 400, 600, 700, np.nan, 1, 2, 3, 4, 5]
```

```python
df = df.replace(df_from, df_to)
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(columns = ["AGRDE_FLAG_NM", "ANSWRR_OC_AREA_NM", "SEXDSTN_FLAG_CD_F", "SEXDSTN_FLAG_CD_M", ])
```

```python
df.dtypes
```

```python
df_sum = df.iloc[:, 1:]
```

```python
df_sum
```

```python
df['TORU_INTRST'] = df_sum.sum(axis=1, numeric_only=True, skipna=True)
```

```python
df = df.loc[:,["HSHLD_INCOME_DGREE_NM", "TORU_INTRST"]]
df = df.dropna()
```

```python
df = df.astype({'HSHLD_INCOME_DGREE_NM':'int64'})
```

```python
df.info()
```

```python
pd.unique(df.HSHLD_INCOME_DGREE_NM)
```

```python
fig ,ax = plt.subplots(1,2)
ax[0].boxplot(df['HSHLD_INCOME_DGREE_NM'])
ax[1].boxplot(df['TORU_INTRST'])
ax[0].set_xticklabels(['HSHLD_INCOME_DGREE_NM'])
ax[1].set_xticklabels(['TORU_INTRST'])
ax[0].set_ylabel('value')
plt.show()
```

```python
from scipy.stats import shapiro

income = df['HSHLD_INCOME_DGREE_NM']
interest = df['TORU_INTRST']
print(shapiro(income))
print(shapiro(interest))

from scipy.stats import levene
print()
print(levene(income,interest))


from scipy.stats import ttest_rel
print(ttest_rel(income,interest))
print()


from scipy.stats import wilcoxon
print(wilcoxon(income,interest))

# 정규성을 가지지 않음
# 정규성을 가지지 않으므로 대응 표본 검정중 비모수 검정인 윌콕슨 부호순위 검정을 진행

#p_value가 3.9644....이므로 "가구소득이 높을수록 여행에 더 큰 관심이 있다"라는 귀무가설을 기각 할 수 없다.
```
