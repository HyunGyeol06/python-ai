```python
import pandas as pd
import datetime

dats_source_path = '../data_set/service_val/'

train_x = pd.read_csv(dats_source_path+"Service_X_train.csv")
train_y = pd.read_csv(dats_source_path+"Service_y_train.csv")
test_x = pd.read_csv(dats_source_path+"Service_X_test.csv")

display(train_x)
display(train_y)
#display(test_x)
```

```python
#train set 으로 사용할 df 중 예측에 사용되지 않는 컬럼
drop_col = ['CustomerId','Surname']

#Test set의 결과 파일에 출력할 CustomerId 저장
test_id = test_x['CustomerId']

#Y의 Train set
y = train_y['Exited']
```

```python
train_x_drop = train_x.drop(columns=drop_col)
test_x_drop = test_x.drop(columns=drop_col)

train_x_dum = pd.get_dummies(train_x_drop)
test_x_dum = pd.get_dummies(test_x_drop)

test_x_dum = test_x_dum[train_x_dum.columns]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, Y_train, Y_valid = train_test_split(train_x_dum, y, test_size = 0.3, random_state = 78, stratify = y)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=78)
rfc.fit(X_train, Y_train)

valid_pred = rfc.predict(X_valid)
valid_pred_proba = rfc.predict_proba(X_valid)[:,1]
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score, precision_score

print('valid accuracy: ', accuracy_score(Y_valid, valid_pred))
print('valid reacll: ', recall_score(Y_valid, valid_pred))
print('valid precision: ', precision_score(Y_valid, valid_pred))
print('valid f1_score: ', f1_score(Y_valid, valid_pred))
print('valid roc_auc_score: ', roc_auc_score(Y_valid, valid_pred_proba))

```

```python
test_pred = rfc.predict(test_x_dum)

pd.DataFrame({"CustomerId": test_id,
              "Exited": test_pred})

```
