```python
import pandas as pd
import numpy as np

train_x = pd.read_csv("carsprice_X_train.csv")
test_x = pd.read_csv("carsprice_X_test.csv")
train_y = pd.read_csv("carsprice_Y_train.csv")


display(train_x)
display(train_y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
stdId = test_x['carID']
drop_col = ["carID"]

train_x_drop = train_x.drop(columns=drop_col)
test_x_drop = test_x.drop(columns=drop_col)
y_train_target = train_y['price']

x_train_dum = pd.get_dummies(train_x_drop)
x_test_dum = pd.get_dummies(test_x_drop)
x_test_dum['model_ M6'] = np.zeros(2672)

x_test_dum = x_test_dum[x_train_dum.columns]

xtr, xt, ytr, yt = train_test_split(x_train_dum, y_train_target)

rf = RandomForestRegressor(random_state = 34)
rf.fit(xtr, ytr)

y_validation_pred = rf.predict(xt)
```

```python
print("validation mse", mean_squared_error(yt, y_validation_pred))
print("validation rmse", np.sqrt(mean_squared_error(yt, y_validation_pred)))
print("validation r2", r2_score(yt, y_validation_pred))
```

```python
test_pred = rf.predict(x_test_dum)

result = pd.DataFrame({"carID" : stdId, "price" : test_pred})

result
```
